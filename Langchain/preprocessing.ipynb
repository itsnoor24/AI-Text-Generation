```python
import pandas as pd
import re
```

```python
df = pd.read_csv("gm_actions.csv")
df.head(20)
```

```python
#drop unneeded columns
df.drop(['releases', 'contributors', 'full_description', 'url', 'version', 'links','status','Unnamed: 0'] , axis=1, inplace=True)
```

```python
# convert to lowercase 
df['name'] = df['name'].str.lower()
df['developer'] = df['developer'].str.lower()
df['short_description'] = df['short_description'].str.lower()
df['categories'] = df['categories'].str.lower()
```

```python
df['short_description'] = df['short_description'].str.replace('[^A-Za-z0-9\s]', ' ', flags=re.UNICODE)
```

```python
df.head(20)
```

```python
# names_to_keep=['code review']
df = df[df['categories'].str.contains('code review', na = False)]
# df = df[df['categories'].isin(names_to_keep)]
df.shape
```
